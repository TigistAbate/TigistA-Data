```python
 import pandas as pd
df = pd.read_csv("student.csv")
```

```python
print(df.head())
```

```python
print(df[df["class"] == "Four"])
```

```python
 df["passed"] = df["mark"] >= 60
print(df)
```

```python
df = df.rename(columns={"mark": "score"})
print(df)
```

```python
df = df.drop(columns=["passed"])
print(df)
```

```python
print(df.groupby("class")["score"].mean())
```

```python
print(df["class"].value_counts())
```

```python
print(df.groupby("gender")["score"].mean())
```

```python
pivot_table = df.pivot_table(values="score", index="class", columns="gender")
print(pivot_table)
```

```python
def get_grade(score):
    if score >= 85:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "D"

df["grade"] = df["score"].apply(get_grade)

print(df)
```

```python
sorted_df = df.sort_values(by="score", ascending=False)
print(sorted_df)
```

```python
df.to_csv("student_with_grade.csv", index=False)

print("File saved successfully.")
```

```python
import matplotlib.pyplot as plt
df.groupby("class")["score"].mean().plot(kind="bar")
plt.title("Average Score by Class")
plt.xlabel("Class")
plt.ylabel("Average Score")
plt.show()
```
